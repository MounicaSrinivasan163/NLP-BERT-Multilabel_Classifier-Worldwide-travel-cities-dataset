# Data Reading

```python
import pandas as pd

df = pd.read_csv("/content/Worldwide Travel Cities Dataset (Ratings and Climate).csv")
df
```

```python
df.info()
```

# Data Preprocessing/Cleaning

```python
ndf=df.iloc[:,10:19]
ndf['description'] = df['short_description']
ndf
```

```python
import pandas as pd

# Suppose df is your DataFrame
rating_cols = ["culture", "adventure", "nature", "beaches",
               "nightlife", "cuisine", "wellness", "urban", "seclusion"]

# Binarize: 1 if >=4 else 0
for col in rating_cols:
    ndf[col] = (ndf[col] >= 4).astype(int)

ndf[rating_cols].head()
```

# Text embedding - BERT

```python
!pip install transformers sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer

# Load a pre-trained Sentence-BERT model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Encode all descriptions into embeddings
embeddings = model.encode(df['short_description'].tolist())
print("Embedding shape:", embeddings.shape)
```

# Train-Model

```python
# Target columns (multi-label),ie. y is a multi-hot vector
y = ndf[rating_cols]

# Features
X = df['short_description']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(embeddings, y, test_size=0.2, random_state=42)

clf = OneVsRestClassifier(LogisticRegression(max_iter=500,class_weight='balanced'))
clf.fit(X_train, y_train)

# Predictions
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred, target_names=rating_cols))
```

# Prediction/Test

```python
new_desc = ["A vibrant city with historic architecture, museums, and lively nightlife near the beach."]
new_embedding = model.encode([new_desc[0]])
pred = clf.predict(new_embedding)

print(dict(zip(rating_cols, pred[0])))
```
